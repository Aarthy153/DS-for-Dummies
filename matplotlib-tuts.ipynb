```python
import matplotlib.pyplot as plt
```

### Basic plot

```python
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.title('Parabola')

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x, y, '-b', label='sine')
plt.plot(x, z, '-r', label='cosine')

plt.title('Simple trigonometric curves')
plt.xlabel('Angle (Radians)')

plt.legend(loc='upper left')

plt.ylim(-1.5, 2.0)

plt.show()
```

### Bar plot

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1]) # first two parameters -> origin, last two -> widths and heights
languages = ['Python', 'Julia', 'R', 'C/C++', 'Java']
developers = [100, 50, 150, 40, 70]

ax.bar(languages, developers)

plt.xlabel('Languages')
plt.ylabel('Number of developers')

plt.show()
```

### Pie chart

```python
fig1 = plt.figure()

ax = fig1.add_axes([0, 0, 1, 1])

languages = ['Python', 'Julia', 'R', 'C/C++', 'Java']
developers = [100, 50, 150, 40, 70]

ax.pie(developers, labels=languages, autopct='%1.1f%%')

plt.show()
```

### Scatter plot

```python
x = np.linspace(0, 10, 50)
y = np.random.random(x.shape) * 5
z = np.random.normal(5, 3, x.shape[0])

fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])

ax.scatter(x, y, color='r')
ax.scatter(x, z, color='g')

plt.show()
```

### 3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)

x = np.sin(z)
y = np.cos(z)

ax.scatter(x, y, z, c=z, cmap='Blues')

plt.show()
```
