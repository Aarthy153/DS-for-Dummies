```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
sns.set_theme()
sns.set_style('darkgrid')
```

### Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

```python
# visualize the dataset

sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

### Scatter plot

```python
# load the iris dataset

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

### Count plot

```python
titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

### Bar chart

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

### Distribution Plot

```python
# house price dataset
from sklearn.datasets import load_boston
```

```python
house_boston = load_boston()
house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
house.head()
```

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

### Correlation

```python
correlation = house.corr()

correlation
```

### Heat Map

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Oranges')
```
